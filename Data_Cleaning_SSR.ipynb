```python
# import dependencies
import pandas as pd
from pathlib import Path
```

```python
# load first dataset
twitter_data = Path("Data/twitter_human_bots_dataset.csv")
twitter_data_df = pd.read_csv(twitter_data)
twitter_data_df.head()
```

```python
list(twitter_data_df.columns)
```

```python
# drop unnecessary columns
twitter_data_df = twitter_data_df.drop(columns=['Unnamed: 0', 'profile_background_image_url', 'profile_image_url', 'average_tweets_per_day',
                                                'account_age_days', 'default_profile_image'])
twitter_data_df.head()
```

```python
# load second dataset
spam_users_data = Path("Data/social_bots_users.csv")
spam_data_df = pd.read_csv(spam_users_data)
spam_data_df.head()
```

```python
list(spam_data_df.columns)
```

```python
# drop unnecessary columns
spam_data_df = spam_data_df.drop(columns=['name','listed_count', 'default_profile_image', 'url', 'time_zone', 'profile_image_url', 'profile_banner_url', 'profile_use_background_image',
                                          'profile_background_image_url_https', 'profile_text_color', 'profile_image_url_https',
                                          'profile_sidebar_border_color', 'profile_background_tile', 'profile_sidebar_fill_color',
                                          'profile_background_image_url', 'profile_background_color', 'profile_link_color', 'utc_offset',
                                          'is_translator', 'follow_request_sent', 'protected', 'notifications', 'contributors_enabled',
                                          'created_at', 'crawled_at', 'updated', 'following'])
spam_data_df.head()
```

```python
# clean up values

twitter_data_df['account_type'] = twitter_data_df['account_type'].replace({'bot': 1, 'human': 0})
twitter_data_df['default_profile'] = twitter_data_df['default_profile'].replace({False: 0, True: 1})
twitter_data_df['geo_enabled'] = twitter_data_df['geo_enabled'].replace({False: 0, True: 1})
twitter_data_df['verified'] = twitter_data_df['verified'].replace({False: 0, True: 1})
twitter_data_df = twitter_data_df[['id', 'screen_name', 'default_profile', 'description', 'created_at', 'geo_enabled', 'lang', 'location', 'statuses_count', 'followers_count', 'friends_count', 'favourites_count', 'verified', 'account_type']]
twitter_data_df.head()
```

```python
# clean up second data
spam_data_df = spam_data_df[['id', 'screen_name', 'default_profile', 'description', 'timestamp', 'geo_enabled', 'lang', 'location', 'statuses_count',
                 'followers_count', 'favourites_count', 'verified', 'test_set_2']]
spam_data_df.fillna(0, inplace=True)
spam_data_df[['default_profile', 'geo_enabled', 'verified']] = spam_data_df[['default_profile', 'geo_enabled', 'verified']].round().astype(int)
spam_data_df.rename(columns={'test_set_2': 'account_type', 'timestamp': 'created_at'}, inplace=True)
spam_data_df.head()
```

```python
# drop location in both
spam_data_df = spam_data_df.drop(columns='location')
twitter_data_df = twitter_data_df.drop(columns='location')
```

```python
# concat datasets
data_df = pd.concat([spam_data_df, twitter_data_df], axis=0)
data_df.drop(columns='friends_count', inplace=True)
data_df.head()
```

```python
# check for NaN values
nan_columns = data_df.isna().any()
columns_with_nan = nan_columns[nan_columns].index.tolist()
print(columns_with_nan)
```

```python
# drop NaN rows
data_df = data_df.dropna()
```

```python
# export dataframe to csv
data_df.to_csv('Data/clean_data.csv', index=False)
```

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

# Load data
data_df = pd.read_csv('/description.csv')

# Feature Engineering
data_df['created_at'] = pd.to_datetime(data_df['created_at'])
data_df['year'] = data_df['created_at'].dt.year
data_df['month'] = data_df['created_at'].dt.month
data_df['day'] = data_df['created_at'].dt.day
data_df['hour'] = data_df['created_at'].dt.hour
data_df['minute'] = data_df['created_at'].dt.minute
data_df.drop(columns=['created_at','screen_name', 'description', 'lang', 'cleaned_description'], inplace=True)

# Split data into features and target
X = data_df
y = data_df['account_type']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

# Handle class imbalance
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Initialize and train the logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train_resampled, y_train_resampled)

# Making predictions on the testing set
y_pred = logreg.predict(X_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Displaying other metrics (precision, recall, F1-score, etc.)
print(classification_report(y_test, y_pred))

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Load data
data_df = pd.read_csv('/description.csv')

# Feature Engineering
data_df['created_at'] = pd.to_datetime(data_df['created_at'])
data_df['year'] = data_df['created_at'].dt.year
data_df['month'] = data_df['created_at'].dt.month
data_df['day'] = data_df['created_at'].dt.day
data_df['hour'] = data_df['created_at'].dt.hour
data_df['minute'] = data_df['created_at'].dt.minute
data_df.drop(columns=['created_at','screen_name', 'description','lang', 'cleaned_description'], inplace=True)

# Split data into features and target
X = data_df.drop(columns=['account_type'])
y = data_df['account_type']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

# Initialize the K-Nearest Neighbors classifier
knn_classifier = KNeighborsClassifier()

# Train the classifier
knn_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("KNN Model Accuracy:", accuracy)
print(classification_report(y_test, y_pred))
```
